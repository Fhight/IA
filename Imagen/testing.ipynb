```python
import cv2 as cv
import numpy as np
from keras.models import load_model

# Load the pre-trained model
model = load_model("./model2-011.model")

# Dictionary for mask results and colors
results = {1: 'without mask', 0: 'mask'}
GR_dict = {0: (0, 0, 255), 1: (0, 255, 0)}

# Open the camera
cap = cv.VideoCapture(0)

# Load the face cascade classifier
rostro = cv.CascadeClassifier('./haarcascade_frontalface_alt.xml')

while True:
    rval, im = cap.read()
    im = cv.flip(im, 1, 1)

    # Detect faces
    faces = rostro.detectMultiScale(im, scaleFactor=1.3, minNeighbors=5)

    for (x, y, w, h) in faces:
        face_img = cv.resize(im[y:y+h, x:x+w], (150, 150))
        normalized = face_img / 255.0
        reshaped = np.reshape(normalized, (1, 150, 150, 3))
        reshaped = np.vstack([reshaped])
        result = model.predict(reshaped)

        label = np.argmax(result, axis=1)[0]

        cv.rectangle(im, (x, y), (x+w, y+h), GR_dict[label], 2)
        cv.rectangle(im, (x, y-40), (x+w, y), GR_dict[label], -1)
        cv.putText(im, results[label], (x, y-10), cv.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)

    cv.imshow('LIVE', im)
    key = cv.waitKey(10)

    if key == 27:
        break

cap.release()
cv.destroyAllWindows()
```
